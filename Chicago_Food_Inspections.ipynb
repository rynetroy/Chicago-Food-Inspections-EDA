# **Part 2: In-Depth Analytical Exploration (Deep Dive)**

After understanding the structure and contents of the dataset, I explored a few key areas more deeply to determine whether this data can answer meaningful public health and operational questions.

## ***1. Risk Level vs Inspection Outcome***

```python
df.groupby('Risk')['Results'].value_counts(normalize=True)
```

### What this does

- This groups inspections by Risk level and then calculates the percentage distribution of each inspection result within that risk category.
- Using normalize=True converts raw counts into proportions, which makes it easier to compare failure rates across Risk 1, Risk 2, and Risk 3 without being biased by how many inspections each group has.

### Why I ran this

- I ran this to see whether the city’s risk classification system actually reflects inspection outcomes.
- If Risk 1 establishments fail more often, that means the risk labeling system is meaningful and aligns with real-world inspection results.
- If the failure rates were similar across all risk levels, it would suggest the classification system might not be very effective.

### What I found

- Across all inspections, about 62% pass, around 6% fail, and the rest are No Entry or Out of Business.
- For Risk 1 (High Risk) establishments:
    - 53% pass
    - 19% fail
    - 16% pass with conditions
- The failure rate is noticeably higher in Risk 1 compared to the overall average.
- This suggests that high-risk establishments are more likely to fail inspections than lower-risk ones.

### Why this matters

This confirms that the dataset supports meaningful policy-level analysis. It shows that risk levels are not random labels — they correlate with inspection outcomes.

That means this dataset can be used to evaluate regulatory targeting, enforcement effectiveness, and public health prioritization.

## ***2. Facility Type & Inspection Performance***

```python
df.groupby('Facility Type')['Results'].value_counts(normalize=True)
```

### What this does

- This groups inspections by Facility Type and calculates the percentage distribution of inspection results within each type.
- Using normalize=True converts the counts into proportions so I can compare performance across facility categories regardless of how many inspections each one has.

### Why I ran this

- I ran this to see whether certain types of establishments perform worse during inspections. If one category consistently shows higher failure rates, that would suggest operational risk patterns.
- However, the output also revealed something important: the dataset contains many granular facility labels, which may need grouping or cleaning before meaningful comparisons.

### What I found

- Some facility types show 100% pass rates, but many of those categories appear only once or twice.
- Small categories like gas stations or nursing homes show extreme proportions (e.g., 33% fail, 50% pass), but this is likely due to very small sample sizes.
- Because there are over 1,500 unique facility types, the data is very fragmented.

This means that raw proportions can be misleading for small groups.

### Why this matters

This investigation shows two things:

- The dataset can support analysis of compliance differences by business type.
- The facility type field likely needs cleaning or grouping (for example, combining similar categories like different types of restaurants).

Without grouping, small categories distort the proportions and reduce reliability.

## ***3. Repeat Inspections & Compliance Behavior***

```python
df.groupby('DBA Name')['Results'].value_counts().head(10)
```

### What this does

- This groups the dataset by business name (DBA Name) and counts how many times each inspection result appears for each business.
- The .head(10) just displays a sample of the grouped results so I can quickly see patterns without printing everything.

### Why I ran this

I ran this to see whether businesses appear multiple times and whether inspection results change across visits.

If businesses only appeared once, then the dataset would only allow static analysis. But since repeat inspections exist, this means I can analyze improvement patterns, repeat violations, or chronic compliance issues.

### What I found

- Some businesses appear multiple times in the dataset, confirming that repeat inspections exist.
- For example, #1 CHOP SUEY appears multiple times with different outcomes (Pass, Fail, Pass with Conditions).
- This suggests that businesses can fail and later pass inspections, indicating changing compliance behavior over time.

This confirms that the dataset supports longitudinal analysis, not just one-time inspection snapshots.

### Why this matters

This investigation shows that the dataset can support deeper behavioral analysis, such as:

- Identifying repeat offenders
- Measuring improvement after failures
- Evaluating enforcement effectiveness

This moves the dataset from simple descriptive analysis to trend and compliance tracking.

## ***4. Inspection Outcomes Over Time***

```python
df['Inspection Date'] = pd.to_datetime(df['Inspection Date'])
df.groupby(df['Inspection Date'].dt.year)['Results'].value_counts(normalize=True)
```

### What this does

First, it converts the inspection date from text into a datetime format. Then it groups inspection results by year and calculates the percentage distribution of outcomes each year.

Then, it groups inspections by year and calculates the percentage distribution of inspection results for each year. Using normalize=True shows proportions instead of raw counts, making it easier to compare trends across time.

### Why I ran this

I ran this to see whether inspection performance improves or declines over time.

If pass rates increase over time, it may suggest better compliance or improved enforcement. If failure rates spike in certain years, it could indicate policy changes, stricter inspections, or broader operational challenges.

### What I found

- In early years like 2010 and 2011, pass rates were around 62–63%.

- Failure rates were significantly higher in earlier years (around 22–24%).

- There are also consistent proportions of “Pass with Conditions” and “Out of Business.”

This suggests that inspection outcomes have varied across years, and that compliance patterns may change over time.

### Why this matters

This confirms the dataset supports trend analysis, not just static summaries.

It allows evaluation of:

- Long-term compliance behavior
- Regulatory effectiveness
- Policy impact over time

That makes the dataset much more powerful for analysis because it supports time-based insights, not just one-year snapshots.
